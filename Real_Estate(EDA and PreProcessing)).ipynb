```python
#to dectection of outliers
for i in df.describe().columns:
    plt.boxplot(x=df[i])
    plt.title(i)
    plt.show()

    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upper_tail=q3+1.3*iqr
    lower_tail=q1-1.3*iqr

    print("q1---",q1)
    print("q3---",q3)  
    print("iqr---",iqr)
    print("upper---",upper_tail)
    print("lower---",lower_tail)
```
